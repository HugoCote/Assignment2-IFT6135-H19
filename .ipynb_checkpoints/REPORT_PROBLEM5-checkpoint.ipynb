### Code taken from ptb-lm.py

```python
# HELPER FUNCTIONS
def _read_words(filename):
    with open(filename, "r") as f:
      return f.read().replace("\n", "<eos>").split()

def _build_vocab(filename):
    data = _read_words(filename)

    counter = collections.Counter(data)
    count_pairs = sorted(counter.items(), key=lambda x: (-x[1], x[0]))

    words, _ = list(zip(*count_pairs))
    word_to_id = dict(zip(words, range(len(words))))
    id_to_word = dict((v, k) for k, v in word_to_id.items())

    return word_to_id, id_to_word

def _file_to_word_ids(filename, word_to_id):
    data = _read_words(filename)
    return [word_to_id[word] for word in data if word in word_to_id]

# Processes the raw data from text files
def ptb_raw_data(data_path=None, prefix="ptb"):
    train_path = os.path.join(data_path, prefix + ".train.txt")
    valid_path = os.path.join(data_path, prefix + ".valid.txt")
    test_path  = os.path.join(data_path, prefix + ".test.txt")

    word_to_id, id_2_word = _build_vocab(train_path)
    train_data = _file_to_word_ids(train_path, word_to_id)
    valid_data = _file_to_word_ids(valid_path, word_to_id)
    test_data  = _file_to_word_ids(test_path , word_to_id)
    return train_data, valid_data, test_data, word_to_id, id_2_word

# Yields minibatches of data
def ptb_iterator(raw_data, batch_size, num_steps):
    raw_data = np.array(raw_data, dtype=np.int32)

    data_len = len(raw_data)
    batch_len = data_len // batch_size
    data = np.zeros([batch_size, batch_len], dtype=np.int32)
    for i in range(batch_size):
        data[i] = raw_data[batch_len * i:batch_len * (i + 1)]

    epoch_size = (batch_len - 1) // num_steps

    if epoch_size == 0:
        raise ValueError("epoch_size == 0, decrease batch_size or num_steps")

    for i in range(epoch_size):
        x = data[:, i*num_steps:(i+1)*num_steps]
        y = data[:, i*num_steps+1:(i+1)*num_steps+1]
        yield (x, y)

def repackage_hidden(h):
    if isinstance(h, Variable):
        return h.detach_()
    else:
        return tuple(repackage_hidden(v) for v in h)

class Batch:
    "Data processing for the transformer. This class adds a mask to the data."
    def __init__(self, x, pad=0):
        self.data = x
        self.mask = self.make_mask(self.data, pad)

    @staticmethod
    def make_mask(data, pad):
        "Create a mask to hide future words."

        def subsequent_mask(size):
            """ helper function for creating the masks. """
            attn_shape = (1, size, size)
            subsequent_mask = np.triu(np.ones(attn_shape), k=1).astype('uint8')
            return torch.from_numpy(subsequent_mask) == 0

        mask = (data != pad).unsqueeze(-2)
        mask = mask & Variable(
            subsequent_mask(data.size(-1)).type_as(mask.data))
        return mask
```

The relevant code for this problem is located in problem5.py -> code_for_51. 
We first instantiate and load a trained model (in problem 4.1) to collect the required data. For the RNN and the GRU we run the model on the validation sequence 2 times, each time computing the loss function at each time step and averaging it across sequence of the same batch, then averaging the result across all the batches. ## Answering 5.1    
To answer this question, we used the models trained in 4.1. The models used achieved the following ppl on the validation set:      
RNN : 149.59 ppl   
GRU : 117.56 ppl    
TRANSFORMER (abbrv. TRANS) : 151.42    

Nous notons les points suivants :    
1) Les courbes des fonctions de coût pour les trois modèles sont ordonnées selon le même ordre que les perplexités : GRU < RNN < TRANS.    
2) La courbe pointillée rejoint la courbe solide pour le RNN beaucoup plus rapidement pour le RNN que pour le GRU, indiquant que les états cachées (qui commencent initialisée et donc non-reliée au contexte des séquences en entrées) du RNN acquierent l'information du contexte des séquences plus rapidement que celles du GRU. Après 25 pas de temps seulement, on voit que la performance moyenne du RNN est la même peu importe s'il a commencé la batch avec des états cachés initialisés ou non. Cela n'est pas le cas pour le GRU pour lequel une seule séquence de longueur 35 n'est pas suffisante pour tout-à-fait acquierir le contexte dont il se sert pour prédire le prochain mot.    
3) De plus, on voit que l'écart entre la courbe pointillée et la courbe solide pour le GRU commence beaucoup plus élevée que pour le RNN indiquant que le GRU accorde plus de poids au contexte (à ses états cachés) que le RNN. Cela explique le GRU est plus pénalisé par l'absence de contexte que le RNN.    
4) Pour le RNN et le GRU, on observe que la perte moyenne en fonction du pas de temps à travers une séquence semble osciller aléatoirement autour d'une valeur constante. Cela est du au fait qu'après avoir acquéri le contexte d'une séquence (via leurs états cachées), ils concervent leurs états cachées pour la batch suivante. Ainsi, pour prédire les premiers mots d'une séquence d'une nouvelle batch, ils ont hérités le contexte pertinent de la batch précédente (la séquence i de la batch j est suivit de la séquence i de la batch j+1 dans le corpus de texte dont les données d'entrainement et de validation sont tirées). Il fallait donc s'attendre à ce que la perte moyenne oscille aléatoirement autour d'une constante car du point de vue du RNN et du GRU l'opération effectué à chaque pas de temps est identique (sauf pour le cas ou les états cachées ont récemment été initialisé.     
5) Pour le RNN et GRU dont les états cachées ont été initalisés à chaque début de batch ainsi que pour le transformer, on observe une perte moyenne plus élevé au début des séquences (<5 premiers pas de temps), cela est du au fait qu'ils ont très peu d'informations pour prédire le prochain mots car leurs prédiction ne sont basées que sur le minuscule contexte (<5 mots) dont ils disposent. De manière à quand même pouvoir faire une bonne prédiction, le TRANS a sans doute du acquérir une connaissance statistique sur la fréquence d'apparition des mots pour les débuts de séquences des données d'entraînement , par exemple en accordant une grande probabilitée au token qui sont très fréquents (i.e. 'the', '<eos>' et '<ukn>'). Par contre, de la façon dont les données d'entraînement ont été construite, le corpus et divisé en séquence de longueur égale (par exemple 35) et les débuts des séquences sont arbitraires car ils ne coincident pas avec le début de phrases naturelles. Il serait intéressant de voir le graphique plus haut pour le cas du TRANS entrainé sur un corpus divisée en (groupe de) phrase naturelle.    
6) Pour le TRANS, on note qu'après les quelques premiers pas de temps, la perte moyenne semble osciller autour d'une certaine valeur comme pour le RNN et GRU. Le fait que la perte moyenne dans le cas du RNN et du TRANS semble plus rapidement se stabiliser une fois qu'une certaine quantité de contexte est atteinte que le GRU semble indiquer que le GRU a appris a plus prendre le contexte éloigné en considération pour faire ses prédictions.

## Answering 5.2         
To answer this question, we used the models trained in 4.1. The models used achieved the following ppl on the validation set:      
RNN : 149.59 ppl   
GRU : 117.56 ppl    
TRANSFORMER (abbrv. TRANS) : 151.42    

Nous notons les points suivants :       
1) La dérivée de la perte pour la prédiction du dernier mot (pas de temps T) est maximale quand elle est prise par rapport à un état cachée au temps T. Ensuite, elle décroit en fonction de la profondeur (layer i est plus proche de l'entrée que layer i+1) de l'état caché et en fonction du temps.     
2) De la façon que le TRANS est programmé, la sortie de la dernière couche du TRANSFORMER STACK pour le temps t n'affecte que la perte au temps t, ansi le gradient est nul pour t!=T dans le cas de la dernière couche (layer 5) et le log de sa norme est -inf pour t!=T et n'apparait pas dans le graphique. On ne voit que les courbes pour les layers 0 à 4.     
3) Les pointillées transparants pour chaque courbes sont allignés horizontalement avec la valeurs en t=1 de la courbe. Ces lignes pointillées servent de repère visuel pour voir l'écart entre la valeur d'une courbe au temps t=1 et t=T.     
4) Pour le RNN, une decroissance lineaire sur un axe des ordonnées en logarithme indique une decroisance exponentielle de la norme du gradient p.r au temps. Ainsi, pour ce modèle on a un indice clair de la présence du phénomène de dissipation (vanishing) du gradient pour les deux couches représentées. Ce phénomène est commun au RNN et est du au fonctionnement du modèle. Dans un RNN, le signal d'un pas de temps au suivant se transmet en passant par une multiplication matrice-vecteur et l'application (potentielle) d'une non-linearité. Lors de la propagation arrière, le gradient est multiplié par cette même matrice et le jacobien de la transformation non-linéaire (qui sera prês de 0 si la non-linéarité choisie est saturée). Lors de la propagation arrière sur un grand nombre de pas de temps, tel que montré dans la question 5 de la partie théorique, ceci peut mené à une décroissance exponentielle de la norme du gradient et donc sa disparition. C'est cela que nous observons.      
5) Pour le GRU, nous observons des courbes qui oscillent autour d'une décroissance linéaire, indiquant également une décroissance exponentielle, mais d'amplitude beaucoup réduit comparativement au RNN. La lenteur de la décroissance permet au modèle de propager le gradient sur un grand nombre de pas de temps et d'apprendre des dépendances long-terme entre les mots. L'architeture GRU (et LSTM dont elle est inspirée) on été conçu pour addresser le problème de disparition de gradient chez un RNN. Lors de la propagation arrière, le gradient passe par un produit de hadamard par un vecteur de 'forget gate' a chaque pas de temps. La multiplication d'un nombre linéaire de scalaire inférieure à 1 mène à une décroissance exponentielle, par contre ces scalaire dans une 'forget gate' ne sont pas affectés par la saturation de non-linéarité comme le jacobien (ils ne sont pas proche de 0 si saturation) et peuvent être tous proche de 1 auquel cas le gradient diminue lentement. On observe effectivement que le taux de décroissance pour le GRU est beaucoup plus petit que pour le RNN. Ceci corobore l'observation faite en 5.1 que le GRU a appris des dépendances plus long-terme que le RNN.     
6) Pour le TRANS, on observe un comportement différent du RNN et GRU.  L'architecture d'un TRANSFORMER est conçu de telle sorte à permettre des connexions directes entre les éléments éloignés d'une séquence. Dans le graphe de calcul de l'architecture TRANSFORMER de profondeur p recevant en entrée une séquence de longueur T, la plus courte distance entre la sortie au temps t=T et un état cachée au temps t<T est tout au plus de O(p) arêtes. Ainsi, durant la propagation arrière, le gradient de la perte au temps t=T par rapport à n'importe quel état caché reçoit la contribution de terme étant multiplié seulement par un petit nombre de matrices. De plus, les connexions résiduelle entre les étages devraient permettent une bonne propagation arrière du gradient d'une étage à la précédente. On observe que la norme du gradient decroit en traversant chaque étage avec à peu près la même magnitude de la décroissance qu'on observe entre le temps t=1 et T=T-1 pour un même étage. On observe également que le gradient decroit rapidement entre le pas de temps t=0 et t=1, puis se stabilise en decroissant très lentement (plus lentement que le RNN et GRU). Contrairement au RNN et GRU, cette décroissance n'est pas du tout causé intrinsequement par l'architecture tel qu'il vient d'être expliqué. Le modèle reçoit l'information de la position d'un élément dans une séquence via le 'positional encoding'. Si le gradient décroit c'est parce que le modèle évalue en moyenne une contribution plus petite de la part des mots éloignées dans le passée sur la prédiction du prochain mot. Cette particularité est apprise et n'est pas forcée. Il pourrait être intéressant d'entraîner un TRANSFORMER sur un corpus de texte de poésie en alexandrin et observer si le modèle apprend une pondération oscillante et périodique en moyenne pour la contribution des mots antérieures pour la prédiction du prochain mot. On pourrait peut-être alors observée un comportement périodique du gradient pour sur un graphique comme celui montré plus haut et comparé le GRU et le transformer.


# Prob 5.3
The sampling functions for this problem are located in model.py, for both the RNN and the GRU we use a method called "generate" that takes seeds and hidden states has input and outputs the following : (sampled sequences, their log-probabilities, hidden states).

We first instantiate and load a trained model (from problem 4.1). In order to initialize the hidden state of the RNN and the GRU we first run the model on few mini-batches. After, we begin to generate samples. In the generation process we prevent the '<ukn>' token from being sampled. This is the only token we exclude. We make this choice in order to produce samples that make more sense. The seeds used to generate the samples are the last words of the previous batch. 

The models were trained with sequences of length 35. We sample for each (2) models 200 sequences of length 35 and 200 sequences of length 70. All those exemple are saved to a file.

Amongst each pack of 200 samples, we only select 10 samples equally spaced from the least probable to the more probable one.  Then for each architecture we are left with 10 short and 10 long samples. For each model we then select 3 best, 3 worst and 3 interesting samples. And comment on them.

#### Comment of the distribution   
We make the following observation : at all times, very few words are absolutely impossible, very few words are very likely, and the bulk of the distribution is a very large number of words that are quite equally likely. When we sample from this type of distribution that was obtain through a softmax we can multiply the input of the softmax by factor. We call this factor the temperature T and call the likely words the 'hot' words. By increasing T, we can increase the likeliness of the 'hot' words and by decreasing it, we can flatten the distribution and make it more probable to sample from the bulk of the distribution which is made from a large number of words that are equally probable.

This notion of temperature in the code appears in the "sample_from_logits" function that is used to sample from this kind of distribution.

### Suite de la discussion     
Les forces du GRU sont qu'il est capable de généré de vrais phrases et qu'il est capable de maintenir un thème général sans répéter sans cesse les mêmes mots. Le défaut du GRU, comme le RNN, est d'attribuer une haute probabilité à certains mots ou certaines combinaisons de mots. On remarque que les 'mauvais' exemplaires sont ceux jugés les plus probables (parmis ceux de leur longueur) par les modèles, ce qui empêche de générer des échantillons via un algorithme 'greedy' qui souhaiterait maximiser la probabilité de généré des phrases cohérentes en prenant seulement les mots que le modèle préfère le plus à chaque étape. Les exemplaires 'mauvais' des deux modèles permettent de voir qu'en faisant cela, les phrases générés sont exagérément répétives, avec les mêmes mots, et ne veulent rien dire. Plutôt, on trouve les meilleurs phrases parmis les échantillons de moindre probabilité (comparer la probabilités exp( -46.1 ) du mauvais échantillon 40 avec la probabilité exp( -218.6 ) du bons échantillon 32).     

Les exemplaires tirés du RNN ne permettent pas d'identifié une particularité distinctive intéressante par rapport au GRU. Ils sont surtout mauvais. On remarque toutefois la présence des mêmes défaut que le RNN, mais amplifié : le modèle aime la répétition et les meilleurs exemplaires sont les moins probables, ce qui rend la génération de bons exemplaires non-triviale.

#### Extra code : sampling from the transformer   
The following code can be used to sample from the transformer. It includes 2 sampling algorithm that only use the model in black-box to sample from it. There is a greedy-beam-search with random sampling algorithm and a truncated-sampler that sample at each step only from the most probable next words.

```python
class SamplerNode:
    # contains sequence of words and the associated probability
    def __init__(self, seq, pro):
        self.seq = seq
        self.pro = pro
    
    def get_values(self):
        return self.seq, self.pro
    
    # define 'lower than' operator
    # comparison is made according to probabilities, not sequences
    def __lt__(self, other):
        return self.pro > other.pro 
    
    # define equality operator, comparison is made according to sequences
    def __eq__(self, other):
        return self.seq == other.seq
    
    def __hash__(self):
        return hash(tuple(self.seq))
    
class BeamSampler:
    # greedy search
    # 'beam_size' is an int (e.g. 5)
    # at each step, search 'beam_size' new child for each current nodes 
    # then pick the 'beam_size' best amongst them
    # repeat until a sequence of desired length has been generated
    
    def __init__(self, net, length, beam_size, trim_size, outp_size, seed, id_2_word, word_to_id) :
        self.net       = net 
        self.length    = length
        self.seed_id   = word_to_id.get(seed)
        self.id_2_word = id_2_word
        self.word_to_id= word_to_id
        self.nodes     = []
        
        self.beam_size = beam_size
        self.trim_size = trim_size
        self.outp_size = outp_size
        
        self.unkn_id = word_to_id['<unk>']
        self.stop_id = word_to_id['<eos>']
        
    def generate_sequence(self):
        
        with torch.no_grad():
            # first step : sample from seed
            self.gen_first_nodes_from_seed()

            # all subsequent step : sample from nodes
            # we have already generated 2 words by now
            for i in range(2,self.length):
                # self.print_nodes()
                self.next_step()
                # trim from time to time, but not close to the end
                if i%5 == 0 and i < self.length - 5 :
                    self.trim_keep_rand()
                    self.remove_redundant_nodes()
            else : # only keep the best
                self.remove_redundant_nodes()
                self.trim_keep_best()
    
    def print_nodes(self):
        # pick the best node
        for i, node in enumerate(sorted(self.nodes)):
            seq, prob = node.get_values()
            print( "node no." , i , " , prob = " , prob , " : \n" )
            print( self.id_seq_2_word_seq(seq) )
    
    def print_readable_nodes(self):
        for i, node in enumerate(sorted(self.nodes)):
            seq, prob = node.get_values()
            print( "\n********************************************\n")
            print( "node no." , i , " , prob = " , prob , " : \n" )
            for word in seq :
                print( self.id_2_word.get(word) , end= " ")
            print( "\n*********************************************\n")
            
    def id_seq_2_word_seq(self, id_seq):
        return [ self.id_2_word.get( i ) for i in id_seq ]
        
    def next_step(self):
        # generete next word suggestion for all current nodes
        # and pick the best amongst them to form the next layer
        # of nodes
        bs              = self.beam_size
        n_node          = len( self.nodes )
        next_nodes      = []
        best_next_nodes = []
        next_nodes_prob = torch.zeros( n_node , bs , dtype=torch.float)
        
        for i,n in enumerate(self.nodes):
            seq, pro = n.get_values()
            next_words, probs = self.sample(seq)
            for j,nw in enumerate(next_words):
                # reject next word suggestion if it has the '<unk>' label
                if nw != self.unkn_id :
                    this_node = seq + [nw.item()] 
                    this_prob = pro + probs[j].item()
                else : 
                    continue
                
                n = SamplerNode( this_node , this_prob )
                best_next_nodes += [n]
                
        self.nodes = best_next_nodes
    
    def trim_keep_best(self):
        os              = self.outp_size
        best_next_nodes = []
        n_nodes  = len( self.nodes )
        logprob = torch.empty( n_nodes )
        for i in range(n_nodes) :
            _, logprob[i] = self.nodes[i].get_values()
            
        values, indices = torch.topk(logprob, os, dim=0)
        for i in indices :
            best_next_nodes += [ self.nodes[i] ]
            
        self.nodes = best_next_nodes
   
    def pick_some_from(self, qt, log_prob_array):
        # qt : number of element to pick randomly with repetition
        # log_prob_array : array of log probabilities
        min_val        = log_prob_array.min()
        log_prob_array = (log_prob_array-2*min_val).abs().sqrt()
        sqr_sum        = log_prob_array.sum()
        log_prob_array = log_prob_array / sqr_sum
        cumulative_dis = torch.cumsum(log_prob_array, dim=0).numpy()
        
        unif_vect = torch.FloatTensor(qt).uniform_(0,1)
        rand_ind  = np.empty(qt, dtype=int) 
        for i in range(qt):
            rand_ind[i] = np.searchsorted(cumulative_dis, unif_vect[i].item() , side='left')
            
        return rand_ind
        
    def trim_keep_rand(self):
        ts              = self.trim_size
        best_next_nodes = []
        n_nodes  = len( self.nodes )
        logprob = torch.empty( n_nodes )
        for i in range(n_nodes) :
            _, logprob[i] = self.nodes[i].get_values()
            
        indices = self.pick_some_from(ts, logprob)
        for i in indices :
            best_next_nodes += [ self.nodes[i] ]
            
        self.nodes = best_next_nodes
        
    def remove_redundant_nodes(self):
        # remove identical nodes 
        # usefull if the current set of nodes as been picked
        # randomly with repetition from another set of nodes
        tmp_set = set()
        for n in self.nodes :
            tmp_set.add(n)
        self.nodes = list( tmp_set )
    
    def prob_dist_from_seq(self, seq) :
        # for transformer :
        batch  = Batch( torch.from_numpy( np.asarray( seq ) ).long().unsqueeze(0).to(device) )
        y = self.net.forward( batch.data, batch.mask )[0,-1,:].cpu()
        return y
        # for the other models :
        # ...
        
    def initialize(self):
        self.nodes = []
        
    def gen_first_nodes_from_seed(self):
        self.initialize()
        y = self.prob_dist_from_seq( [self.seed_id] )
        probs , top_id = torch.topk(y, self.beam_size + 1, dim=-1)
        
        for i,j in enumerate(top_id) :
            if j.item() == self.unkn_id : # exclude '<unk>' label
                continue
            if len(self.nodes) >= self.beam_size : # 
                break
            n = SamplerNode( [self.seed_id, j.item()] , probs[i] )
            self.nodes += [n]
            
    def sample(self, node):
        # return self.beam_size (e.g. 3) suggestions for the next word, 
        # given node as current sequence
        with torch.no_grad() :
            y = self.prob_dist_from_seq(node)
            probs , top_id = torch.topk(y, self.beam_size, dim=-1)
            
        # the model return log probabilities, we convert them to real prob by applying exp.
        return top_id , probs
        
```

```python
seed      = '<eos>'
beam_size = 3   # number of nodes to explore at each step
trim_size = 10  # number of nodes picked at random with repetition during trimming
outp_size = 20  # number of nodes to keep at the end
length    = 70  # length of the output sentence, measured by the number of token (i.e. words)

bsampler = BeamSampler( net.to(device) , length, beam_size, trim_size, outp_size, seed, id_2_word, word_to_id )
```

### Approximation of a true sampler :

```python
class TrueSampler:
    # output a random sequence of fixed lenght by sampling each consecutive
    # word from true distribution truncated to the most probable suggestions
    # trunc_search : nb of most probable suggestions we sample from at each step
    
    def __init__(self, net, length, trunc_search, seed, id_2_word, word_to_id) :
        self.net          = net 
        self.length       = length
        self.id_2_word    = id_2_word
        self.word_to_id   = word_to_id
        self.trunc_search = trunc_search
        self.initialize()
        
        self.real_pro   = 0
        self.trun_pro   = 0
        
        self.unkn_id = word_to_id['<unk>']

    def generate_sequence(self):
        with torch.no_grad():
            # all step : sample from current seq
            for i in range(1,self.length):
                self.next_step()
    
    def print_readable(self):
        seq = self.sequence
        print( "\n********************************************\n")
        print( "current node, real prob = " , self.real_pro.item() , "truncated prob = " , self.trun_pro.item() , " : \n" )
        for word in seq :
            print( self.id_2_word.get(word) , end= " ")
        print( "\n*********************************************\n")
        
    def next_step(self):
        # generete next word suggestion for current node
        seq       = self.sequence
        next_word = self.sample(seq)
        self.sequence += [next_word]
   
    def pick_one_from(self, log_prob_array):
        # pick one element randomly from :
        # log_prob_array : array of log probabilities
        max_val        = log_prob_array.max()
        log_prob_array = (log_prob_array-max_val).exp()  # adjust scaling
        pro_sum        = log_prob_array.sum()
        log_prob_array = log_prob_array / pro_sum
        cumulative_dis = torch.cumsum(log_prob_array, dim=0).numpy()
        
        unif_01  = torch.FloatTensor(1).uniform_(0,1)
        rand_ind = np.searchsorted(cumulative_dis, unif_01.item() , side='left')
            
        return rand_ind, log_prob_array[rand_ind].log()
    
    def prob_dist_from_seq(self, seq) :
        # for transformer :
        batch  = Batch( torch.from_numpy( np.asarray( seq ) ).long().unsqueeze(0).to(device) )
        y = self.net.forward( batch.data, batch.mask )[0,-1,:].cpu()
        return y
        # for the other models :
        # ...
        
    def initialize(self):
        self.sequence   = [ word_to_id[seed] ]
            
    def sample(self, node): 
        # sample, but exclude '<unk>' id
        y = self.prob_dist_from_seq(node)
        probs , top_id = torch.topk(y, self.trunc_search, dim=-1)
        
        next_id = self.unkn_id
        while next_id == self.unkn_id :
            rand_ind, trunc_prob = self.pick_one_from( probs )
            next_id = top_id[rand_ind].item()
            
        # update (true and truncated) probability of the current sequence
        self.real_pro += probs[rand_ind]
        self.trun_pro += trunc_prob
        
        return next_id
        
```
